##### ROUGE Scores for News18

```python
import pandas as pd
import itertools

def ngram_overlap(reference, generated, n=1):
    """Calculate Precision, Recall, and F1-score for ROUGE-N (unigram, bigram)"""
    ref_tokens = reference.split()
    gen_tokens = generated.split()

    ref_ngrams = list(zip(*[ref_tokens[i:] for i in range(n)]))
    gen_ngrams = list(zip(*[gen_tokens[i:] for i in range(n)]))

    overlap = len(set(ref_ngrams) & set(gen_ngrams))
    precision = overlap / max(len(gen_ngrams), 1)  # Avoid division by zero
    recall = overlap / max(len(ref_ngrams), 1)
    if precision + recall == 0:
        f1 = 0
    else:
        f1 = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1

def lcs_length(X, Y):
    """Helper function to calculate Longest Common Subsequence (LCS) length"""
    m, n = len(X), len(Y)
    dp = [[0] * (n+1) for _ in range(m+1)]
    
    for i, j in itertools.product(range(1, m+1), range(1, n+1)):
        if X[i-1] == Y[j-1]:
            dp[i][j] = dp[i-1][j-1] + 1
        else:
            dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    
    return dp[m][n]

def rouge_l(reference, generated):
    """Calculate Precision, Recall, and F1-score for ROUGE-L (LCS-based)"""
    ref_tokens = reference.split()
    gen_tokens = generated.split()
    lcs = lcs_length(ref_tokens, gen_tokens)

    precision = lcs / max(len(gen_tokens), 1)
    recall = lcs / max(len(ref_tokens), 1)
    if precision + recall == 0:
        f1 = 0
    else:
        f1 = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1

def calculate_rouge_scores(df):
    """Compute ROUGE scores for the dataset"""
    for prompt in ["Prompt 1", "Prompt 2", "Prompt 3"]:
        rouge1_p, rouge1_r, rouge1_f1 = [], [], []
        rouge2_p, rouge2_r, rouge2_f1 = [], [], []
        rougeL_p, rougeL_r, rougeL_f1 = [], [], []

        for i, row in df.iterrows():
            ref_summary = row["summary"]
            gen_summary = row[prompt]

            p1, r1, f1_1 = ngram_overlap(ref_summary, gen_summary, n=1)  # ROUGE-1
            p2, r2, f1_2 = ngram_overlap(ref_summary, gen_summary, n=2)  # ROUGE-2
            pl, rl, f1_l = rouge_l(ref_summary, gen_summary)  # ROUGE-L

            rouge1_p.append(p1); rouge1_r.append(r1); rouge1_f1.append(f1_1)
            rouge2_p.append(p2); rouge2_r.append(r2); rouge2_f1.append(f1_2)
            rougeL_p.append(pl); rougeL_r.append(rl); rougeL_f1.append(f1_l)

        df[f"{prompt} ROUGE-1 Precision"] = rouge1_p
        df[f"{prompt} ROUGE-1 Recall"] = rouge1_r
        df[f"{prompt} ROUGE-1 F1"] = rouge1_f1

        df[f"{prompt} ROUGE-2 Precision"] = rouge2_p
        df[f"{prompt} ROUGE-2 Recall"] = rouge2_r
        df[f"{prompt} ROUGE-2 F1"] = rouge2_f1

        df[f"{prompt} ROUGE-L Precision"] = rougeL_p
        df[f"{prompt} ROUGE-L Recall"] = rougeL_r
        df[f"{prompt} ROUGE-L F1"] = rougeL_f1

    return df

# Load the dataset
df = pd.read_csv("summarized_data_news18_cleaned.csv")  # Update with your dataset path
df = calculate_rouge_scores(df)

# Save the updated dataset
df.to_csv("summarized_data_news18_with_rouge.csv", index=False)
print("ROUGE scores added and saved successfully!")
df.head(10)
```

```python
import pandas as pd
import itertools

def ngram_overlap(reference, generated, n=1):
    """Calculate Precision, Recall, and F1-score for ROUGE-N (unigram, bigram)"""
    ref_tokens = reference.split()
    gen_tokens = generated.split()

    ref_ngrams = list(zip(*[ref_tokens[i:] for i in range(n)]))
    gen_ngrams = list(zip(*[gen_tokens[i:] for i in range(n)]))

    overlap = len(set(ref_ngrams) & set(gen_ngrams))
    precision = overlap / max(len(gen_ngrams), 1)  # Avoid division by zero
    recall = overlap / max(len(ref_ngrams), 1)
    if precision + recall == 0:
        f1 = 0
    else:
        f1 = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1

def lcs_length(X, Y):
    """Helper function to calculate Longest Common Subsequence (LCS) length"""
    m, n = len(X), len(Y)
    dp = [[0] * (n+1) for _ in range(m+1)]
    
    for i, j in itertools.product(range(1, m+1), range(1, n+1)):
        if X[i-1] == Y[j-1]:
            dp[i][j] = dp[i-1][j-1] + 1
        else:
            dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    
    return dp[m][n]

def rouge_l(reference, generated):
    """Calculate Precision, Recall, and F1-score for ROUGE-L (LCS-based)"""
    ref_tokens = reference.split()
    gen_tokens = generated.split()
    lcs = lcs_length(ref_tokens, gen_tokens)

    precision = lcs / max(len(gen_tokens), 1)
    recall = lcs / max(len(ref_tokens), 1)
    if precision + recall == 0:
        f1 = 0
    else:
        f1 = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1

def calculate_rouge_scores(df):
    """Compute ROUGE scores for the dataset"""
    for prompt in ["Prompt 1", "Prompt 2", "Prompt 3"]:
        rouge1_p, rouge1_r, rouge1_f1 = [], [], []
        rouge2_p, rouge2_r, rouge2_f1 = [], [], []
        rougeL_p, rougeL_r, rougeL_f1 = [], [], []

        for i, row in df.iterrows():
            ref_summary = row["summary"]
            gen_summary = row[prompt]

            p1, r1, f1_1 = ngram_overlap(ref_summary, gen_summary, n=1)  # ROUGE-1
            p2, r2, f1_2 = ngram_overlap(ref_summary, gen_summary, n=2)  # ROUGE-2
            pl, rl, f1_l = rouge_l(ref_summary, gen_summary)  # ROUGE-L

            rouge1_p.append(p1); rouge1_r.append(r1); rouge1_f1.append(f1_1)
            rouge2_p.append(p2); rouge2_r.append(r2); rouge2_f1.append(f1_2)
            rougeL_p.append(pl); rougeL_r.append(rl); rougeL_f1.append(f1_l)

        df[f"{prompt} ROUGE-1 Precision"] = rouge1_p
        df[f"{prompt} ROUGE-1 Recall"] = rouge1_r
        df[f"{prompt} ROUGE-1 F1"] = rouge1_f1

        df[f"{prompt} ROUGE-2 Precision"] = rouge2_p
        df[f"{prompt} ROUGE-2 Recall"] = rouge2_r
        df[f"{prompt} ROUGE-2 F1"] = rouge2_f1

        df[f"{prompt} ROUGE-L Precision"] = rougeL_p
        df[f"{prompt} ROUGE-L Recall"] = rougeL_r
        df[f"{prompt} ROUGE-L F1"] = rougeL_f1

    return df

# Load the dataset
df = pd.read_csv("summarized_data_nenow_cleaned.csv")  # Update with your dataset path
df = calculate_rouge_scores(df)

# Save the updated dataset
df.to_csv("summarized_data_nenow_with_rouge.csv", index=False)
print("ROUGE scores added and saved successfully!")
df.head(10)
```
